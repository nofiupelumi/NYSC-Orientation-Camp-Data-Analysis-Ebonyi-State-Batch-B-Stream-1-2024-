```python
# NYSC Ebonyi Data Analysis for 2024 Batch B Stream 1

# Step-by-step code for Colab to replicate charts and tables in the official report

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set Seaborn style for charts
sns.set(style="whitegrid")

# Step 1: Load the data (assuming you've saved the cleaned data as CSV)
# You can replace this with the actual CSV once available
data = pd.read_csv("/content/nysc_ebonyi_2024.csv")

# Step 2: Plot - State of Origin Distribution
plt.figure(figsize=(12, 6))
data['State of Origin'].value_counts().sort_values(ascending=False).plot(kind='bar')
plt.title('Distribution of Corps Members by State of Origin')
plt.xlabel('State')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=75)
plt.tight_layout()
plt.show()

```

```python
# Step 3: Gender Distribution
plt.figure(figsize=(6, 4))
data['Gender'].value_counts().plot(kind='bar', color=['skyblue', 'lightcoral'])
plt.title('Gender Distribution of Respondents')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Step 4: Age Distribution Histogram
plt.figure(figsize=(8, 5))
data['Age'].plot(kind='hist', bins=10, color='mediumseagreen', edgecolor='black')
plt.title('Age Distribution of Corps Members')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Step 5: Religion Pie Chart
religion_counts = data['Religion'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(religion_counts, labels=religion_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("pastel"))
plt.title('Religion of Respondents')
plt.tight_layout()
plt.show()

```

```python
# Step 6: Marital Status Pie Chart
marital_counts = data['Marital Status'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(marital_counts, labels=marital_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Marital Status of Respondents')
plt.tight_layout()
plt.show()

```

```python
# Step 7: Satisfaction with Kits (Bar Chart)
kits = data['Satisfied with Kits Issued'].value_counts()
plt.figure(figsize=(6, 4))
kits.plot(kind='bar', color=['mediumblue', 'salmon'])
plt.title('Satisfaction with Kits Issued')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

```

```python
# Step 8: Feeding Satisfaction (Grouped Bar Chart)
feeding = data[['Food Quantity', 'Food Quality', 'Involvement in Prep', 'Mode of Serving']].apply(pd.Series.value_counts).T.fillna(0)
feeding[['Very Satisfied', 'Satisfied', 'Fairly Satisfied', 'Not Satisfied']].plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Feeding Satisfaction Levels')
plt.ylabel('Respondent Count')
plt.xlabel('Feeding Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
# Step 9: Camp Discipline Chart
discipline_cols = ['NYSC Officials', 'Soldiers', 'Police/NSCDC', "Man O' War", 'Others']

camp_discipline = data[discipline_cols].apply(pd.Series.value_counts).T.fillna(0)
camp_discipline[['Very Disciplined', 'Disciplined', 'Not Disciplined']].plot(kind='bar', stacked=True, figsize=(10, 5))
plt.title('Assessment of Camp Discipline')
plt.xlabel('Group')
plt.ylabel('Number of Respondents')
plt.tight_layout()
plt.show()


```

```python
# Step 10: Save summary table
summary = data.describe(include='all')
summary.to_csv("nysc_summary_statistics.csv")
```
